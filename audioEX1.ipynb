```python
import os
import librosa
import numpy as np
import matplotlib.pylab as plt
import glob

# glob 라이브러리를 활용해 wav 확장자를 가진 모든 파일 directory 저장 
voice_list = glob.glob('voiceSample/*.wav')
```

```python
# 음성을 저장하는 공간
for i in range(len(voice_list)):
    voiceWav = voice_list[i]
    (voice_file_dir, voice_file_id) = os.path.split(voiceWav)

    y, sr = librosa.load(voiceWav, sr=16000)
    # y축으로 사용할 시간 생성
    time = np.linspace(0, len(y)/sr, len(y)) 

    fig, ax1 = plt.subplots()
    ax1.plot(time, y, color = 'b', label='speech waveform')
    ax1.set_ylabel("Amplitude")
    ax1.set_xlabel("Time [s]") 
    plt.title(voice_file_id)
    # plt.savefig(voice_file_id+'.png')
    plt.show()
    if i == 2 : break
```
